```python
%%writefile calculos.py
import random

def menor(a, b):
  return a if a < b else b

def divide(a: float, b: float) -> float:
  if b == 0:
      raise ValueError("No se puede dividir por 0")
  return a / b

def numero_aleatorio_1_10():
  resultado = random.randint(1, 10)
  if resultado == 4:
    resultado = 5
  return resultado
```

```python
# comprobando un resultado completo
%%writefile test_calculos.py
from calculos import menor

def test_menor():
    assert menor(2, 3) == 2
```

```python
# Pruebas parametrizadas
%%writefile test_calculos.py

import pytest
from calculos import menor

@pytest.mark.parametrize("a,b,esperado", [
    (1, 1, 1),
    (0, 5, 2),
    (-1, 1, -1),
])

def test_menor_parametrizada(a, b, esperado):
    assert menor(a, b) == esperado
```

```python
# Probando excepciones
%%writefile test_calculos.py

from calculos import divide
import pytest

def test_divide_ok():
    assert divide(10, 2) == 5

def test_divide_por_cero():
    with pytest.raises(ValueError):
        divide(10, 0)
```

```python
# Probando con bucles
%%writefile test_calculos.py

from calculos import numero_aleatorio_1_10

def test_numero_aleatorio_1_10_esta_en_rango():
    for i in range(100):
        n = numero_aleatorio_1_10()
        assert 1 <= n <= 9, f"Error en el número aleatorio {i}"
```

```python
!pytest -q
```

```python
# Comprobando que se dan todos los resultados
%%writefile test_calculos.py

from calculos import numero_aleatorio_1_10

def test_numero_aleatorio_genera_todos_los_valores():
    resultados = set()

    for _ in range(1000):  # muchas repeticiones
        n = numero_aleatorio_1_10()
        resultados.add(n)

    assert resultados == set(range(1, 11)), (
        f"No se generaron todos los valores. "
        f"Obtenidos: {sorted(resultados)}"
    )
```
