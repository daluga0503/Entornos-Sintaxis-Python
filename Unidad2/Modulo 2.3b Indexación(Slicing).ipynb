```python
'''
Podemos tratar un string como un array de caracteres
<var>[a:b] => caracteres de <var> desde la posición a (incluida) hasta la b(no incluida)
ej:
  indice:        0123456789
  string:        hola_mundo
'''
c = "hola_mundo"
print(c[0])
print(c[1:3])
print(c[:5])
print(c[4:])
print(c[:])
```

```python
'''
También podemos usar índices negativos para recorrer cadenas delde el final
  indice:        0123456789
  string:        hola mundo
                 ^        ^
  indice -:     -10       -1

'''
c = "hola_mundo"
print(c[-1])
print(c[-10])
print(c[-2:])
```

```python
c = "hola_mundo"
l = len(c)
print(l)
for i in range(0,l):
  print(c[i])
```

```python
c = "hola_mundo"
for i in range(len(c)):
  print(c[i:])
```

```python
'''
  También podemos especificar la cantidad de salto
'''
c = "hola_mundo"
print(c[0:-1:2])
print(c[0::3])
```
