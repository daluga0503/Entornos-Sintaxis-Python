```python
a=0b101 & 0b111
print(a)
print(f"{a:b}")

a=0b101 | 0b111
print(f"{a:b}")
```

```python
# Operaciones lógicas vs bit a bit
a = 6   # 0b110
b = 1   # 0b001

print("a & b =", a & b)
print("a and b =", a and b)
print("a | b =", a | b)
print("a or b =", a or b)
print("0 or b =", 0 or b)
print("0 and b =", 0 and b)
```

```python
# Las operaciones booleanas usan "cortocircuito" para no tener que evaluar no que no es necesario
def f():
    print("ejecuto f()")
    return True

print(f() and False)
print(False and f())
```

```python
# Los operadores >> y << permiten desplazamiento de bits
# Por ejemplo 0b1000 >> 2 = 0b0010
var = 17 # 10001
var_right = var >> 1
var_left = var << 2
print(f"{var:b}", f"{var_left:b}", f"{var_right:b}")
print(var, var_left, var_right)
```

```python
# En el operador negación bit a bit se usa representación en complemento a 2
# ~n = -(n + 1)
print (~0)
print (~1)
print (~5)
```

```python
# Si queremos usar el operador ~ para invertir sin complemento a 2 necesitamos usar máscaras
n = 0b101
bits = 3

mask = (1 << bits) - 1   # 0b111
r = ~n & mask            # invierto y recorto a 3 bits

print(f"{r:b}")   # 010
```
