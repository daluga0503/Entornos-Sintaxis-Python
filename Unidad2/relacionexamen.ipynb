```python
# 1. Realizar un programa que muestre un menú con las siguientes opciones:
# 1 Suma
# 2 Resta
# 3 Multiplicación
# 5 División
# 0 Salir
# A continuación solicitará 2 números y mostrará el resultado de la elegida.
# El programa se repetirá continuamente hasta que el usuario seleccione la opción de salir

def menu():
    salir = False
    print('Menu de Opciones:')
    print('1. Suma')
    print('2. Resta')
    print('3. Multiplicación')
    print('4. División')
    print('0. Salir')
    opcion = input('Selecciona una opción entre 0-4: ')
    while not salir:
        try:
            opcion = int(opcion)
            num1 = float(input('Introduce el primer número: '))
            num2 = float(input('Introduce el segundo número: '))
            match opcion:
                case 1:
                    print(f'El resultado de la suma es: {num1 + num2}')
                    menu()
                case 2:
                    print(f'El resultado de la resta es: {num1 - num2}')
                    menu()
                case 3:
                    print(f'El resultado de la multiplicación es: {num1 * num2}')
                    menu()
                case 4:
                    if num2 != 0:
                        print(f'El resultado de la división es: {num1 / num2}')
                    else:
                        print('Error: No se puede dividir entre cero.')
                    menu()
                case 0:
                    salir = True
                    print('Saliendo del programa. ¡Hasta luego!')
                case _:
                    print('Opción no válida. Por favor, selecciona una opción entre 0-4.')
                    menu()
        except ValueError as ve:
            print(f'Error: Debes inreoducir un número válido, que sea positivo y entero: {ve}')
            menu()
        except Exception as e:
            print(f'Error inesperado: {e}')
            menu()
menu()

```

```python
# 2. Realizar un programa que convierta grados Celsius a Fahrenheit
def celsius_to_farenheit(celsius):
    farenheit = (celsius * 1.8) + 32
    return farenheit

try:
    celsius = float(input('Introduce la temperatura en grados Cº: '))
except ValueError as ve:
    print(f'Error: Debes introducir un número válido: {ve}')
except Exception as e:
    print(f'Error inesperado: {e}')
else:
    farenheit = celsius_to_farenheit(celsius)
    print(f'La temperatura en grados Fº es: {farenheit}')
```

```python
# 9. Pide una cadena de caracteres por teclado e imprímela en 2 líneas dividida por la mitad. No
# se pueden usar bucles. Por ejemplo, la cadena "casa" daría como resultado
# ca
# sa
# Y la cadena "remar" daría
# rem
# ar

cadena = input('Introduce una cadena de caracteres: ')

if len(cadena) % 2 == 0:
    mitad = len(cadena) // 2
else:
    mitad = (len(cadena) // 2) + 1
print(cadena[:mitad])
print(cadena[mitad:])
```

```python
# 10. Toma una frase y crea una nueva cadena donde las letras en índice par estén en minúscula y as de índice impar en mayúscula.
frase = input('Introduce una frase: ')
nueva_frase = ''.join([letra.lower() if i % 2 == 0 else letra.upper() for i, letra in frase])
print(nueva_frase)
```

```python
# 1. Aplicación del Operador OR ( | )
# Ejercicio 22: Aplica el operador OR lógico a nivel de bit (|) entre los siguientes dos números binarios: 0b11001010 y 0b10101100. 
# Muestra el resultado final en formato binario.
numero1 = 0b11001010
numero2 = 0b10101100
resultado = numero1 | numero2
print(bin(resultado))

```
