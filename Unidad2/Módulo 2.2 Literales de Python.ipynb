```python
# Literales que parecen idénticos pero no lo son
print("2")
print(2)
```

```python
# Podemos ver la diferencia con el operador de suma
print("2" + "2")
print(2 + 2)
```

```python
# Los literales enteros son números naturales que no pueden tener parte decimal
print(1232)
```

```python
# Los enteros no se escriben con separadores de miles
print(1,232)
print(1.232)
```

```python
# Por claridad se permite el guion bajo como separador
# En realidad el separador se descarta
print(1_232)
# Y se puede colocar en cualquier lugar
print(1_2_3_2)
```

```python
# También podemos usar el signo "-" para enteros negativos
print(-1232)
# Y el signo "+" opcional para los positivos
print(+1232)
```

```python
# Números binarios
print(0b1011)
```

```python
# Númmeros octales
print(0o123)
print(0o15 + 0o16)
```

```python
# Números hexadecimales
print(0xff)
```

```python
# Números flotantes
print(2.5)
print(.5)
print(4.)
```

```python
# Números flotantes en notación científica
# la e indica potencia de 10 (base y exponente)
print(3e8)
print(6.62607E-34)
print(12e2)
print(12e-2)
```

```python
# Python elige por defecto la notación más corta del número
print(0.0000000000000000000001)
print(2e-5)
```

```python
# Cadenas de caracteres
print("Me gusta \"Monty Python\"")
print('Me gusta "Monty Python"')
```

```python
# Booleanos
print(True)
print(False)
# Esto es una cadena, no un booleano
print("True")
```

```python
# El resultado de un operador de comparación es un booleano
print(1 > 2)
print(1 < 2)
```
