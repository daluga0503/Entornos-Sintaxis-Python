**¿Qué son las Variables?**

Elementos del lenguaje que permiten almacenar el resultado de una expresión, es decir, un valor, una operación o el resultado de ejecutar una función.

Tienen 3 componentes:
- Un nombre
- Un tipo
- Un valor

**Reglas sobre los nombres de variable**

- Debe de estar compuesto por MAYÚSCULAS, minúsculas, dígitos, y el carácter _ (guion bajo).
- Debe comenzar con una letra.
- El carácter guion bajo es considerado una letra.
- Las mayúsculas y minúsculas se tratan de forma distinta (un poco diferente que en el mundo real - Alicia y ALICIA son el mismo nombre, pero en Python son dos nombres de variable distintos, subsecuentemente, son dos variables diferentes).
- No pueden ser igual a alguna de las palabras reservadas de Python (se explicará más de esto pronto).

Ejemplos de nombres correctos: MiVariable, i, t34, Tasa_Cambio, contador, dias_para_navidad,Adiós_Señora, ElNombreEsTanLargoQueSeCometeranErroresConEl, _

Ejemplos de nombres incorrectos:
- **10t** (no comienza con una letra)
- **Tasa Cambio** (contiene un espacio)

```python
# En python no se declaran las variables, se crean cuando le asignamos valor por primera vez
nombre="Antonio"
print("Hola", nombre)
```

```python
# Podemos usar variables en lugar de literales
var1 = 12
var2 = 14
var3 = var1 + var2
print(var3)
```

```python
# Tambien se pueden usar en la misma linea como operando y como resultado de la asignación
var1 = 12
var1 = 2*var1
print(var1)
```

```python
# Python es un lenguaje fuertemente tipado
# Si usamos una variable con un tipo incorrecto se produce un error
v_entera = 12
v_cadena = "alan"
print(v_entera + v_cadena)
```

```python
# El tipo de una variable es dinámico, puede cambiar en cada asignación
variable = 12
print(type(variable))
variable = "alan"
print(type(variable))
```

```python
# Ejemplo: el teorema de pitágoras
# El cuadrado de la hipótenusa es igual a la suma de los cuadrados de los catetos
cateto1 = 3.0
cateto2 = 4.0
hipotenusa = (cateto1**2 + cateto2**2) ** 0.5
print(hipotenusa)
```

```python
# Operadores abreviados
# En lugar de <variable> = <variable> <operador> <expresión>
# usamos <variable> <operador>= <expresión>

x = 4
x += 5
print(x)

x *= 2
print(x)

x +=2 * 4
print(x)
```
