**Definiciones**
1. Una expresión es una combinación de valores (o variables, operadores, llamadas a funciones, aprenderás de ello pronto) las cuales son evaluadas y dan como resultado un valor, por ejemplo, 1 + 2.

2. Los operadores son símbolos especiales o palabras clave que son capaces de operar en los valores y realizar operaciones matemáticas, por ejemplo, el * multiplica dos valores: x * y.

```python
# Operadores básicos

# Suma
print("2+4 =", 2+4)
# Resta
print("4-2 =", 4-2)
# Multiplicación
print("4*2 =", 4*2)
# División
print("5/2 =", 5/2)
# División entera
print("5//2 =", 5//2)
# Módulo o residuo (resto de la división)
print("5%2 =", 5%2)
# Potencia
print("2**4 =", 2**4)
```

```python
# Mezclando tipos en operaciones
print (2 ** 3)
print (2 ** 3.)
print(6 / 3)
print(6 / 3.)
print(6//3)
print(6//3.5)
```

```python
# Operadores unarios y binarios
print(-3)
print(2*-5)
print(2*(-5))
```

```python
# Orden de ejecución de los operadores
# Normalmente de derecha a izquierda
'''
Orden de prioridad
1 + - (unario)
2 **
3 * / // %
4 + - (binario)
'''
print(2 + 3 * 5)
print((2 + 3) * 5)
print(2 * 3 % 5)
print(2 % 3 * 5)


# Las potencias de derecha a izquierda
print(2 ** 2 ** 3)
print((2 ** 2) ** 3)


# En caso de duda usa los paréntesis
#print((2 + 3) * 5)

```
