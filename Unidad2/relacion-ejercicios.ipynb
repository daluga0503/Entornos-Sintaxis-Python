```python

# Ejercicio 1 1. Realizar un programa que muestre un menú con las siguientes opciones:
# 1 Suma
# 2 Resta
# 3 Multiplicación
# 5 División
# 0 Salir

def pedir_validar_num():
    try:
        num1 = int(input('Introduce un número: '))
        num2 = int(input('Introduce un número: '))
        return num1, num2
    except Exception as e:
        raise ValueError(f'Se deben meter números, no caracteres. {e}')
    
def menu():
    print('1 Suma')
    print('2 Resta')
    print('3 Multiplicación')
    print('4 División')
    print('0 Salir')

def sumar(x, y):
    return x + y

def restar(x, y):
    return x - y

def multiplicar(x, y):
    return x * y

def dividir(x, y):
    return x / y

salir = False

while not salir:
    menu()
    try:
        opcion = int(input('Introduce una opción del menú (número): '))
    except Exception as e:
        print(f'La opción introducida debe de ser un número: {e}')
    match opcion:
        case 1:
            num1, num2 = pedir_validar_num()
            print(f'La suma de {num1} + {num2} = {sumar(num1, num2)}')
        case 2:
            num1, num2 = pedir_validar_num()
            print(f'La resta de {num1} - {num2} = {restar(num1, num2)}')
        case 3:
            num1, num2 = pedir_validar_num()
            print(f'La multiplicación de {num1} * {num2} = {multiplicar(num1, num2)}')
        case 4:
            num1, num2 = pedir_validar_num()
            print(f'La división de {num1} / {num2} = {dividir(num1, num2)}')
        case 0:
            salir = True
            break
        case _:
            print('Introduce una opción correcta.')


```

```python
# Ejericicio 7 Escribe una expresión que determine si una contraseña es válida cuando:
# • Tiene más de 8 caracteres
# • Contiene la letra "@"
# • No contiene espacios

password = input('Introduce una contraseña que contenga: + de 8 caracteres, contenga \"@\" y no contenga espacios')
if len(password) == 0:
    print('No has introducido ningún valor.')
elif len(password) <= 8 and '@' not in password and ' ' in password:
    print('Recuerda que la contraseña debe de tener + de 8 caractares, un \"@\" y no puede contener espacios')
elif len(password)<=8:
    print('La contraseña debe de tener más de 8 caracteres.')
elif '@' not in password:
    print('La contraseña debe de tener el caracter \"@\"')
elif ' ' in password:
    print('La contraseña no debe de contener ningún espacio.')
else:
    print('La contraseña es correcta')
```

```python
# Ejercicio 9. Pide una cadena de caracteres por teclado e imprímela en 2 líneas dividida por la mitad. No se pueden usar bucles. Por ejemplo, la cadena "casa" daría como resultado
#ca
#sa
#Y la cadena "remar" daría
#rem
#ar

cadena = input('Introduce una cadena: ')
if len(cadena) % 2 == 0:
    print(cadena[:len(cadena)//2])
    print(cadena[len(cadena)//2:])
else:
    print(cadena[:len(cadena)//2+1])
    print(cadena[len(cadena)//2+1:])
```

```python
# Ejercicio 10. Toma una frase y crea una nueva cadena donde las letras en índice par estén en minúscula y as de índice impar en mayúscula.
cadena = input('Introduce una cadena: ')
resultado = ''
for i in range(len(cadena)):
    resultado += cadena[i].lower() if i % 2 == 0 else cadena[i].upper()
print(resultado)
```
