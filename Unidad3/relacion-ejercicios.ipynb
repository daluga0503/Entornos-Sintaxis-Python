```python
# Ejercicio 16
def menu():
    print('===== Menú de opciones =====')
    print('n. Añade un libro.')
    print('p. Extrae un libro de la pila.')
    print('l. Listar los libros de la pila.')
    print('b. Borra todos los libros de la pila.')
    print('x. Salir del programa.')

    print('============================')

pila = []

def agregar_libro(pila):
    titulo_libro = input('Introduce el titulo del libro: ')
    pila.append(titulo_libro)
    print(f'El libro "{titulo_libro}" ha sido añadido a la pila.')

def extraer_libro(pila):
    if len(pila)==0:
        print('La pila está vacía. No hay libros para extraer.')
        return
    
    libro_extraido = pila.pop()
    print(f'El libro "{libro_extraido}" ha sido extraído de la pila.')

def listar_libros(pila):
    if len(pila) == 0:
        print('No hay libros en la pila.')
        return
    print('Libros en la pila:')
    for libro in reversed(pila):
        print(f'-{libro}')

def borrar_libros(pila):
    if len(pila) == 0:
        print('La pila ya está vacía.')
        return
    pila.clear()
    print('Todos los libros han sido borrados de la pila.')

while True:
    menu()

    opcion = input('Selecciona una opción: ').lower()

    match opcion:
        case 'n':
            agregar_libro(pila)
        case 'p':
            extraer_libro(pila)
        case 'l':
            listar_libros(pila)
        case 'b':
            borrar_libros(pila)
        case 'x':
            print('Saliendo del programa...')
            break
        case _:
            print('Opción no válida. Por favor, selecciona una opción del menú.')
```

```python
# Ejercicio 17
from collections import deque
def menu():
    print('===== Menú de opciones =====')
    print('n. Añadir una cita.')
    print('a. Atender una cita')
    print('l. Listar citas.')
    print('b. Borrar todas las citas pendientes.')
    print('============================')

cola = deque()

def agregar_cita(cola):
    nombre_paciente = input('Introduce el nombre del paciente: ')
    cola.append(nombre_paciente)
    print(f'La cita para "{nombre_paciente}" ha sido añadida a la cola.')

def atender_cita(cola):
    if len(cola) == 0:
        print('No hay citas pendientes.')
        return
    paciente_atendido = cola.popleft()
    print(f'La cita para "{paciente_atendido}" ha sido atendida.')

def listar_citas(cola):
    if len(cola) == 0:
        print('No hay citas pendientes.')
        return
    print('Citas pendientes:')
    for paciente in cola:
        print(f'- {paciente}')

def borrar_citas(cola):
    if len(cola) == 0:
        print('No hay citas pendientes para borrar.')
        return
    cola.clear()
    print('Todas las citas pendientes han sido eliminados.')

while True:
    menu()
    opcion = input('Introduce una opción: ').lower()

    match opcion:
        case 'n':
            agregar_cita(cola)
        case 'a':
            atender_cita(cola)
        case 'l':
            listar_citas(cola)
        case 'b':
            borrar_citas(cola)
        case 'x':
            print('Saliendo del programa...')
            break
        case __:
            print('Opción inválida. Introduce uan opción del menú.')
```
