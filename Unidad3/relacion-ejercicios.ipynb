```python
# Ejercicio 1
lista_10_num_enteros = [i for i in range(10)]
print(f'Suma de los valores de la lista: {sum(lista_10_num_enteros)}')
```

```python
# Ejercicio 2
lista_nombres = ['Daniel', 'Diego', 'Ismael', 'Francisco', 'Ana']
contador_vocal = 0
for nombre in lista_nombres:
    if nombre[0] in ['A', 'E', 'I', 'O', 'U']:
        contador_vocal +=1

print(f'{contador_vocal} nombres empiezan por vocal.')
```

```python
# Ejercicio 3
lista_numeros =  [i for i in range(1,21)]
lista_pares = [par for par in lista_numeros if par%2==0]
print(lista_numeros)
print(lista_pares)
```

```python
# Ejercicio 4
lista_bits = [0,0,0,1,1,0,1]
lista_sin_0 = [valor for valor in lista_bits if valor != 0]
print(lista_sin_0)
```

```python
# Ejercicio 5
lista_elementos = [i for i in range(1, 6)]
longitud = len(lista_elementos)
ultimo = lista_elementos[longitud -1]
print(lista_elementos)
for i in range(longitud -2, -1, -1):
    lista_elementos[i+1] = lista_elementos[i]

lista_elementos[0] = ultimo
print(lista_elementos)
```

```python
# Ejercicio 6
lista_multiplo_3 = [valor for valor in range(1, 100) if valor%3 == 0]
print(lista_multiplo_3)
```

```python
# Ejercicio 7
lista_num_repetidos = [1,1,2,6,3,4,4,5]
lista_sin_repetidos = []
vistos = []

for valor in lista_num_repetidos:
    if valor not in vistos:
        lista_sin_repetidos.append(valor)
        vistos.append(valor)

print(lista_sin_repetidos)

```

```python
# Ejercicio 8
lista = [i for i in range(1,11)]
sublista = lista[3:8]
print(lista)
print(sublista)
```

```python
# Ejercicio 9
lista_ultimos_5_elementos = lista[-5:]
print(lista_ultimos_5_elementos)
```

```python
# Ejercicio 10
lista_3_en_3 = [lista[i] for i in range(1, len(lista), 3)] 
print(lista_3_en_3)
```
