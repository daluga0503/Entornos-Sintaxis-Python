a: El médico atiende al paciente al que le toca, se muestra por pantalla en nombre del

l: Muestra un listado de las citas pendientes

```python
# 15
expresion_1 = '(10 + (5 * 2))' # Balanceada
expresion_2 = '(2 * (3 + 4))' # Balanceada
expresion_3 = '(10 + 5) * 2)' # No 
expresion_4 = '(8 + 2 * (5 - 1)' # No
expresion_5 = '5 + 2) * (3' # No
expresion_6 = '())(()'

expresiones = [expresion_1, expresion_2, expresion_3, expresion_4, expresion_5, expresion_6]
def balanceo(expresion):
    pila = []
    for char in expresion:
        if char == '(':
            pila.append(char)
        elif char == ')':
            if len(pila) == 0:
                return 'No'
            pila.pop()
    if len(pila) == 0:
        return 'SI'
    else:
        return 'No'

for expresion in expresiones:
    resultado = balanceo(expresion)
    print(f"{expresion}: {resultado}")
```

```python
# 16
pila_libros = []

def menu():
    print('n. Añadir un libro a la pila.')
    print('p. Extraer el libro.')
    print('l. Mostrar listado de libros.')
    print('b. Borrar todos los libros.')
    print('x. Salir de la aplicación.')

while True:
    menu()
    opcion = input('Seleccione una opción: ')
    match opcion:
        case 'n':
            libro = input('Ingrese el libro a añadir: ')
            pila_libros.append(libro)
        case 'p':
            if len(pila_libros) > 0:
                libro_extraido = pila_libros.pop()
                print(f'Libro extraído: {libro_extraido}')
            else:
                print('La pila de libros está vacía.')
        case 'l':
            if len(pila_libros) > 0:
                for libro in pila_libros:
                    print(libro)
            else:
                print('La pila de libros está vacía.')
        case 'b':
            pila_libros.clear()
        case 'x':
            break
        case _:
            print('Opción no válida. Intente de nuevo.')
```

```python
# 17. Crear una aplicación que simule un sistema de turnos (médico): se atiende al primero y se 
# añade uno nuevo si llega. Se mostrará el siguiente menú:
# n: Pide por pantalla el nombre del paciente y lo añade como nueva cita
# a: El médico atiende al paciente al que le toca, se muestra por pantalla en nombre del paciente y se saca de las citas pendientes
# l: Muestra un listado de las citas pendientes
# b: Borra todas las citas pendientes (pidiendo confirmación)
# x: Sale de la aplicación

from collections import deque
turnos = deque()
def menu_turnos():
    print('n. Añadir un paciente a la lista de turnos.')
    print('a. Atender al siguiente paciente.')
    print('l. Mostrar listado de pacientes pendientes.')
    print('b. Borrar todas las citas pendientes.')
    print('x. Salir de la aplicación.')

while True:
    menu_turnos()
    opcion = input('Seleccione una opción: ')
    match opcion:
        case 'n':
            paciente = input('Ingrese el nombre del paciente a añadir: ')
            turnos.append(paciente)
        case 'a':
            if turnos:
                paciente_atendido = turnos.popleft()
                print(f'Paciente atendido: {paciente_atendido}')
            else:
                print('No hay pacientes pendientes.')
        case 'l':
            if turnos:
                for paciente in turnos:
                    print(paciente)
            else:
                print('No hay pacientes pendientes.')
        case 'b':
            turnos.clear()
        case 'x':
            break
        case _:
            print('Opción no válida. Intente de nuevo.')
```
